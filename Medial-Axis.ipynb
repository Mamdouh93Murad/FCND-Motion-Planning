```python
# TODO: Your start and goal location defined above
# will not necessarily be on the skeleton so you
# must first identify the nearest cell on the 
# skeleton to start and goal

def find_start_goal(skel, start, goal):
    # TODO: find start and goal on skeleton
    # Some useful functions might be:
        # np.nonzero()
        cells = skel.nonzero()
        # np.transpose()
        cells = np.transpose(cells)
        # np.linalg.norm()
        min_start = np.linalg.norm(np.array(start) - np.array(cells), axis=1).argmin()
        #closest_start = cells[min_start]
        min_goal = np.linalg.norm(np.array(goal) - np.array(cells), axis=1).argmin()
        #closest_goal = cells[min_goal]
        # np.argmin()
        near_start = cells[min_start]
        near_goal = cells[min_goal]
        return near_start, near_goal

skel_start, skel_goal = find_start_goal(skeleton, start_ne, goal_ne)

print(start_ne, goal_ne)
print(skel_start, skel_goal)
```

```python
def heuristic_func(position, goal_position):
    cost = abs(goal_position[0] - position[0]) + abs(goal_position[1] - position[1])
    return cost
```

```python
def actions_to_path(actions, start):
    actual_path = []
    current_position = start
    for action in actions:
        current_position = (current_position[0] + action.delta[0], current_position[1] + action.delta[1])
        actual_path.append(current_position)
    return actual_path
```

```python
path = actions_to_path(actions, skel_start)
path2 = actions_to_path(actions2, start_ne)
```
